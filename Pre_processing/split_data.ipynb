```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/Delhi_raw.csv')
```

```python
import warnings


warnings.filterwarnings("ignore")  # avoid printing out absolute paths
```

```python
df.isna().sum()
```

```python
df_new = df.sort_values(['from_date', 'to_date'])

# Set time index
df_new = df_new.set_index(['from_date', 'to_date'])

interpolated_df = df_new.interpolate(
    method='linear'
)

# Reset index if needed
interpolated_df = interpolated_df.reset_index()

interpolated_df.head()
```

```python
interpolated_df["from_date"] = pd.to_datetime(interpolated_df['from_date'])
interpolated_df['Date'] = interpolated_df["from_date"].dt.date
interpolated_df.head()
```

```python
interpolated_df.isna().sum()
```

```python
interpolated_df.shape
```

```python
df_new = interpolated_df
```

```python
df_new['time_idx'] = range(1, len(df_new) + 1)
```

```python
print(len(df_new['time_idx'].unique()))
```

```python
df_new.head()
```

```python
df_new.columns
```

```python
df_new['city'] = "Delhi"
df_new.head()
```

```python
df_new = df_new.sort_values("Date")
```

```python
df_new.to_csv("Delhi_finaldataset_2017.csv")
```

```python
df_new["Date"] = pd.to_datetime(df_new["Date"])
train_data = df_new[df_new["Date"] < "2023-01-01"].copy()
test_data  = df_new[df_new["Date"] >= "2023-01-01"].copy()
```
